#  Исследование результатов А/В-теста и поиск инсайтов

SollmaFin — международное мобильное приложение для онлайн-торговли финансовыми активами. Пользователи могут инвестировать деньги в акции, валюту, криптовалюту, биржевые фонды (ETF) и другие активы. Целевая аудитория приложения — начинающие инвесторы. Получить доступ к финансовым рынкам можно со смартфона либо через веб-версию. 

Приложение ориентировано на рынок Латинской Америки с акцентом на четыре страны: Мексику, Бразилию, Колумбию и Аргентину — там быстро растёт интерес к финтеху и инвестициям, а экономики очень разные. 

Появилась гипотеза о недостаточной финансовой грамотности пользователей: они не понимают, как работать с активами разной степени риска. Клиенты покупают активы с высоким риском и теряют деньги, что снижает вовлечённость.
Команда продукта решила обновить онбординг пользователей и добавить в него детальную информацию о различиях финансовых активов и связанных с ними рисках. Однако появилось опасение, что углублённый онбординг отпугнёт клиентов от пополнения депозита и более рискованных вложений. 

Чтобы разобраться, команда решила провести A/B-эксперимент. 

### А/В-эксперимент

В эксперименте участвовали новые пользователи, которые зарегистрировались в приложении со 2 по 15 июня 2025 года. 
Пользователей случайным образом разделили на две равные группы: 
- Контрольная группа проходила стандартный онбординг без обязательного обучения.
- Тестовая группа проходила обновлённый онбординг с подробной информацией об активах и связанных с ними рисках. 

После разделения активность пользователей анализировали в течение недели.

Цель эксперимента — оценить, как обновлённый онбординг влияет на поведение пользователей и их дальнейшую инвестиционную активность.

При внедрении новой фичи команда продукта выдвинула такие гипотезы:
- Гипотеза роста: обучающий онбординг помогает пользователям лучше понимать принципы инвестирования, поэтому они будут чаще открывать второй депозит.
- Гипотеза риска: информация о возможных потерях и высоких рисках отпугнёт некоторых новичков, особенно самых осторожных, что снизит конверсию в первый депозит.
- Дополнительная гипотеза: после нового онбординга пользователи, которые выбрали высокорискованные активы, будут чаще, чем раньше, возвращаться и открывать второй депозит. При старом онбординге пользователи часто покупали активы с высоким риском без понимания последствий. Это приводило к потерям и оттоку после первого депозита. 

Чтобы увидеть полную картину, недостаточно следить за метриками конверсии. Команда выбрала для эксперимента такой набор метрик: 
- Ключевая метрика — средняя сумма всех депозитов на одного пользователя (включая тех, кто установил приложение или открыл веб-версию).
- Барьерная метрика — конверсия из регистрации в первый депозит.
- Вспомогательная метрика 1 — конверсия из первого депозита во второй.
- Вспомогательная метрика 2 — средняя сумма всех депозитов на пользователя, который открыл хотя бы один депозит.

Ожидалось, что ключевая и барьерная метрики не упадут, а вспомогательные покажут значительный рост.

```python
# определяем выбросы
def detect_outliers(df: pd.DataFrame, column: str):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

outliers_hist = detect_outliers(df_sollmafin_hist, 'amount')
outliers_abt = detect_outliers(df_sollmafin_abt, 'amount')
print('Количество выбросов в df_sollmafin_hist:', len(outliers_hist))
print('Количество выбросов в df_sollmafin_abt:', len(outliers_abt))

# визуализация распределения признака 'amount'
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(df_sollmafin_hist['amount'], bins=30, color='blue', alpha=0.7, edgecolor='black')
plt.title('Распределение amount в df_sollmafin_hist')
plt.xlabel('amount')
plt.ylabel('Частота')
plt.subplot(1, 2, 2)
plt.hist(df_sollmafin_abt['amount'], bins=30, color='orange', alpha=0.7, edgecolor='black')
plt.title('Распределение amount в df_sollmafin_abt')
plt.xlabel('amount')
plt.ylabel('Частота')
plt.tight_layout()
plt.show()
```

**Задача 2.2.** Анализ воронок событий.
- Построим классическую и пошаговую воронки;
- Охарактеризуем пользовательский путь в приложении и определим узкие места воронки. 


```python
# определим последовательность шагов воронки

steps = [
    'install / open_web', 
    'introduction', 
    'registration', 
    'main_page', 
    'onboarding_complete', 
    'first_deposit', 
    'asset_purchase', 
    'second_deposit'
]
```

```python
# считаем воронки
funnel_df = df_sollmafin_hist[df_sollmafin_hist['event_name'].isin(steps)]
funnel_df = funnel_df.groupby('event_name')['user_id'].nunique().reindex(steps).reset_index(name = 'count_users')
funnel_df['conversion_%'] = funnel_df['count_users'] / funnel_df['count_users'].loc[0] * 100
funnel_df['conversion_step_by_step_%'] = funnel_df['count_users'] / funnel_df['count_users'].shift(1) * 100
funnel_df
```

```python
# уберём два первых шага - "install / open_web" и "introduction", они имеют 100% конверсии
# и выполняются всеми пользователями по умолчанию
steps_filter = [
    'install / open_web', 
    'introduction'
]

# визуализируем
funnel_df_plot = funnel_df[~funnel_df['event_name'].isin(steps_filter)]
fig, axes = plt.subplots(2, 1, figsize=(12, 10))

# первый график
bars1 = axes[0].bar(funnel_df_plot['event_name'], funnel_df_plot['conversion_%'])
axes[0].set_title('Классическая конверсия')
axes[0].grid(axis='y', linestyle='--', alpha=0.7)
# добавляем значения на столбцы
# с помощью plt.text
for bar in bars1:
    height = bar.get_height()
    axes[0].text(bar.get_x() + bar.get_width() / 2, height, f'{height:.1f}%', ha='center', va='bottom', fontsize=9)

# второй график
bars2 = axes[1].bar(funnel_df_plot['event_name'], funnel_df_plot['conversion_step_by_step_%'])
axes[1].set_title('Конверсия step_by_step')
axes[1].grid(axis='y', linestyle='--', alpha=0.7)
# добавляем значения на столбцы
# с помощью plt.text
for bar in bars2:
    height = bar.get_height()
    axes[1].text(bar.get_x() + bar.get_width() / 2, height, f'{height:.1f}%', ha='center', va='bottom', fontsize=9)

plt.subplots_adjust(hspace=0.2)

fig.text(0.5, 0.01, 'Шаги', ha='center')
fig.text(0.08, 0.5, 'Конверсия %', va='center', rotation='vertical')

plt.show()
```

**Задача 2.3.** Влияние уровня риска актива на открытие второго депозита.
- Изучим, как уровень риска первого актива влияет на конверсию во второй депозит.

**Задача 3.3.** Влияние новой фичи на конверсию во второй депозит с учётом уровня риска купленного актива.
- Проанализируем, как категория риска купленного актива влияет на вероятность открытия второго депозита в тестовой и контрольной группах. 
- Сравним результаты и оцените, есть ли статистически значимая разница между группами. 

```python
# проверим, как уровни риска влияют на открытие второго депозита пользователями

# функция расчета конверсии по уровням риска
def risk_level(df: pd.DataFrame):
    # отфильтруем пользователей по депозитам
    first_deposit = df[df['event_name'] == 'first_deposit'][['user_id']]
    second_deposit = df[df['event_name'] == 'second_deposit'][['user_id']]

    # оставляем пользователей, которые совершили первый депозит и купили активы
    asset_first_deposit = df[df['user_id'].isin(first_deposit['user_id'])].reset_index()
    asset_second_deposit = df[df['user_id'].isin(second_deposit['user_id'])].reset_index()
    
    # группируем по уровням риска
    asset_first_deposit = asset_first_deposit.groupby('risk_level')['user_id'].nunique().reset_index(name = 'first')
    asset_second_deposit = asset_second_deposit.groupby('risk_level')['user_id'].nunique().reset_index(name = 'second')

    # объединяем 
    risk_level_merge = asset_first_deposit.merge(asset_second_deposit, on = 'risk_level')

    # сортируем по уровням риска в нужном порядке
    risk_order = ['low', 'medium', 'high']
    risk_level_merge['risk_level'] = pd.Categorical(
        risk_level_merge['risk_level'], 
        categories=risk_order, 
        ordered=True
    )
    risk_level_merge['conversion'] = (risk_level_merge['second'] / risk_level_merge['first'] * 100).round(2)
    
    return risk_level_merge.sort_values('risk_level')
```

```python
# подготовка данных контрольной группы
control_risk_level = risk_level(group_control)

# подготовка данных тестовой группы
test_risk_level = risk_level(group_test)
```

```python
# объединим на одном графике для наглядности
risk_comparison = pd.DataFrame({
    'risk_level': test_risk_level['risk_level'],
    'control': control_risk_level['conversion'],
    'test': test_risk_level['conversion']    
})

ax = risk_comparison.plot(x='risk_level', kind='bar', figsize=(8, 6), colormap='tab10', edgecolor = 'black')

plt.title('Конверсия во второй депозит по уровням риска - %')
plt.xlabel('Уровни риска', fontsize = 12)
plt.ylabel('Конверсия')
plt.legend(shadow = True)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

После нового онбординга пользователи, которые выбрали высокорискованные активы, будут чаще, чем раньше, возвращаться и открывать второй депозит.     
 - `Н0` - Конверсия во второй депозит тестовой группы равна или мнеьше конверсии контрольной группы.    
 - `Н1` - Конверсия во второй депозит тестовой группы выше конверсии контрольной группы. 

```python
# применим поправку Бонферрони 
alpha_bonferroni = alpha / 3  # 3 сравнения (low, medium, high)

z_risk_result = []

# проведем z-тест для каждой пары групп и оценим статистическую значимость
for risk in risk_comparison['risk_level']:
    control = control_risk_level[control_risk_level['risk_level'] == risk]
    test = test_risk_level[test_risk_level['risk_level'] == risk]
 
    count_control = control['first'].values[0]
    count_test = test['first'].values[0]
    
    success_control = control['second'].values[0]
    success_test = test['second'].values[0]

    success = [success_test, success_control]
    count = [count_test, count_control]    
    stat, pval = proportions_ztest(success, count, alternative='larger')
    z_risk_result.append({
        'risk_level': risk,
        'z_stat': stat,
        'p_value': f'{pval:.6f}',
        'conversion_control': success_control / count_control * 100,
        'conversion_test': success_test / count_test * 100,
        'significant': f'{"Нулевая гипотеза отклонена" if pval <= alpha_bonferroni else "Альтернативная гипотеза отклонена"}'
    })
    
z_risk_result_df = pd.DataFrame(z_risk_result)
z_risk_result_df
```

Тестовая группа по всем группам риска показала высокий рост конверсии. Сильнее всего это отразилось в конверсии высокого уровня риска. 
- `low` +7.6%
- `medium` +2.5%
- `high` +18.3%

Z - тест статистической значимости показал, что существенные изменения есть только в 3 группе риска `high`.     
Дополнительная гипотеза подтверждена. После введения нового онбординга пользователи стали чаще возвращаться и делать второй депозит, в высокорискованных активах.

**Задача 3.4.** Анализ метрик А/В-эксперимента. Проанализируем такие метрики:
- Ключевая метрика — средняя сумма всех депозитов на одного пользователя (включая тех, кто установил приложение или открыл веб-версию).
- Барьерная метрика — конверсия из регистрации в первый депозит.
- Вспомогательная метрика 1 — конверсия из первого депозита во второй.
- Вспомогательная метрика 2 — средняя сумма всех депозитов на пользователя, который открыл хотя бы один депозит.

Для каждой метрики выведем: 
- значения в контрольной и тестовой группах, 
- абсолютные и относительные изменения значений в тестовой группе по сравнению с контрольной.

После этого проверим статистическую значимость различий метрик между группами эксперимента. 

Для ключевой метрики также изучим:
- накопленную динамику изменения по дням эксперимента для каждой группы, 
- стабильность p-value во время эксперимента.

Сделаем промежуточный вывод. 


```python
all_users_abt = df_sollmafin_abt.groupby('user_id').agg(
    group = ('group', 'first'),
    registration = ('event_name', lambda x: 1 if (x == 'registration').any() else 0),
    first_deposit = ('event_name', lambda x: 1 if (x == 'first_deposit').any() else 0),
    second_deposit = ('event_name', lambda x: 1 if (x == 'second_deposit').any() else 0),
    is_paying = ('amount', lambda x: 1 if (x > 0).any() else 0),
    total_spent = ('amount', 'sum')
).reset_index()

all_users_abt_grouped = all_users_abt.groupby('group').agg(
    count_users = ('user_id', 'nunique'),
    registration = ('registration', 'sum'),
    first_deposit = ('first_deposit', 'sum'),
    second_deposit = ('second_deposit', 'sum'),
    total_spent = ('total_spent', 'sum'),
    count_paying = ('is_paying', 'sum')
)
```

```python
all_users_abt_grouped['arpu'] = (all_users_abt_grouped['total_spent'] / all_users_abt_grouped['count_users']).round(2)
all_users_abt_grouped['conv_reg_to_first_deposit'] = \
    (all_users_abt_grouped['first_deposit'] / all_users_abt_grouped['registration'] * 100).round(2)
all_users_abt_grouped['conv_first_to_second_deposit'] = \
    (all_users_abt_grouped['second_deposit'] / all_users_abt_grouped['first_deposit'] * 100).round(2)
all_users_abt_grouped['arpu_paying'] = (all_users_abt_grouped['total_spent'] / all_users_abt_grouped['count_paying']).round(2)
all_users_abt_grouped
```

```python
# проведем тест статистичккой значимости для всех метрик
alpha = 0.05

# функция для проведения t-теста и оценки значимости
def perform_t_test(metric: str, equal_var: bool = False, alternative: str = 'less'):
    control_values = all_users_abt[all_users_abt['group'] == 'control'][metric]
    test_values = all_users_abt[all_users_abt['group'] == 'test'][metric]
    stat, p_value = ttest_ind(test_values, control_values, equal_var=equal_var, alternative=alternative)
    return stat, p_value

# функция для проведения z-теста для конверсий
def perform_z_test(count_a, nobs_a, count_b, nobs_b, alternative: str):
    count = [count_a, count_b]
    nobs = [nobs_a, nobs_b]
    stat, p_value = proportions_ztest(count, nobs, alternative=alternative)
    return stat, p_value
```

### Сформулируем гипотезы.
`Гипотеза роста`: обучающий онбординг помогает пользователям лучше понимать принципы инвестирования, поэтому они будут чаще открывать второй депозит.
 - `Н0` - Конверсия во второй депозит тестовой группы равна конверсии контрольной группы.   
 - `Н1` - Конверсия во второй депозит тестовой группы не равна конверсии контрольной группы.    

`Гипотеза риска`: информация о возможных потерях и высоких рисках отпугнёт некоторых новичков, особенно самых осторожных, что снизит конверсию в первый депозит.     
 - `Н0` - Конверсия в первый депозит тестовой группы больше или равна конверсии контрольной группы.    
 - `Н1` - Конверсия в первый депозит тестовой группы ниже конверсии контрольной группы.     

Ключевая метрика ARPU вырастет за счет возвращения пользователей
 - `Н0` - ARPU в тестовой группе равна ARPU в контрольной группе.   
 - `Н1` - ARPU в тестовой группе выше, чем в контрольной.   

ARPU среди платящих пользователей вырастет, после прохождения обучения и увеличения лояльности пользователя    
 - `Н0` - ARRPU в тестовой группе равна ARRPU в контрльной группе.  
 - `Н1` - ARRPU в тестовой группе выше ARRPU контрольной.    

```python
# значения заранее для читаемости
control_arpu = all_users_abt_grouped.loc['control', 'arpu']
test_arpu = all_users_abt_grouped.loc['test', 'arpu']
arpu_rel_diff = test_arpu / control_arpu - 1
arpu_abs_diff = test_arpu - control_arpu

control_conv_reg = all_users_abt_grouped.loc['control', 'conv_reg_to_first_deposit']
test_conv_reg = all_users_abt_grouped.loc['test', 'conv_reg_to_first_deposit']
conv_reg_rel_diff = test_conv_reg / control_conv_reg - 1
conv_reg_abs_diff = test_conv_reg - control_conv_reg

control_conv_first = all_users_abt_grouped.loc['control', 'conv_first_to_second_deposit']
test_conv_first = all_users_abt_grouped.loc['test', 'conv_first_to_second_deposit']
conv_first_rel_diff = test_conv_first / control_conv_first - 1
conv_first_abs_diff = test_conv_first - control_conv_first

control_arpu_paying = all_users_abt_grouped.loc['control', 'arpu_paying']
test_arpu_paying = all_users_abt_grouped.loc['test', 'arpu_paying']
arpu_paying_rel_diff = test_arpu_paying / control_arpu_paying - 1
arpu_paying_abs_diff = test_arpu_paying - control_arpu_paying

tstat, pval_arppu = ttest_ind(
    all_users_abt[(all_users_abt['group'] == 'test') & (all_users_abt['is_paying'] == 1)]['total_spent'],
    all_users_abt[(all_users_abt['group'] == 'control') & (all_users_abt['is_paying'] == 1)]['total_spent'],
    equal_var=False,
    alternative='greater'
)

print(
    f'''
      Вспомогательная метрика (конверсия из первого депозита во второй депозит):
      Control: {control_conv_first}%
      Test: {test_conv_first}%
      Относительная разница: {conv_first_rel_diff:.2%}
      Абсолютная разница: {conv_first_abs_diff:.2f}%
      Гипотеза роста, статус: {"Нулевая гипотеза отклонена" if perform_z_test(
            count_a=all_users_abt_grouped.loc['control', 'second_deposit'],
            nobs_a=all_users_abt_grouped.loc['control', 'first_deposit'],
            count_b=all_users_abt_grouped.loc['test', 'second_deposit'],
            nobs_b=all_users_abt_grouped.loc['test', 'first_deposit'],
            alternative='two-sided'
      )[1] < 0.05 else "Нулевая гипотеза не отклонена"}
      p-значение: {perform_z_test(
            count_a=all_users_abt_grouped.loc['control', 'second_deposit'],
            nobs_a=all_users_abt_grouped.loc['control', 'first_deposit'],
            count_b=all_users_abt_grouped.loc['test', 'second_deposit'],
            nobs_b=all_users_abt_grouped.loc['test', 'first_deposit'],
            alternative='two-sided'
      )[1]:.6f}
      
      Барьерная метрика (конверсия из регистрации в первый депозит):
      Control: {control_conv_reg}%
      Test: {test_conv_reg}%
      Относительная разница: {conv_reg_rel_diff:.2%}
      Абсолютная разница: {conv_reg_abs_diff:.2f}%
      Гипотеза риска, статус: {"Нулевая гипотеза отклонена" if perform_z_test(
            count_a=all_users_abt_grouped.loc['control', 'first_deposit'],
            nobs_a=all_users_abt_grouped.loc['control', 'registration'],
            count_b=all_users_abt_grouped.loc['test', 'first_deposit'],
            nobs_b=all_users_abt_grouped.loc['test', 'registration'],
            alternative = 'smaller'
      )[1] < 0.05 else "Нулевая гипотеза не отклонена"}
      p-значение: {perform_z_test(
            count_a=all_users_abt_grouped.loc['control', 'first_deposit'],
            nobs_a=all_users_abt_grouped.loc['control', 'registration'],
            count_b=all_users_abt_grouped.loc['test', 'first_deposit'],
            nobs_b=all_users_abt_grouped.loc['test', 'registration'],
            alternative='smaller'
      )[1]:.6f}
      
      Средний доход на пользователя (ARPU):
      Control: {control_arpu}
      Test: {test_arpu}
      Относительная разница: {arpu_rel_diff:.2%}
      Абсолютная разница: {arpu_abs_diff:.2f} у.е.
      Статус: {"Нулевая гипотеза отклонена" if perform_t_test('total_spent')[1] < alpha
            else "Нулевая гипотеза не отклонена"}
      p-значение: {perform_t_test('total_spent')[1]:.6f}
      
      Вспомогательная метрика (ARPPU):
      Control: {control_arpu_paying}
      Test: {test_arpu_paying}
      Относительная разница: {arpu_paying_rel_diff:.2%}
      Абсолютная разница: {arpu_paying_abs_diff:.2f} у.е.
      Статус: {"Нулевая гипотеза отклонена" if pval_arppu < alpha else "Нулевая гипотеза не отклонена"}
      p-значение: {pval_arppu:.6f}
      '''
)
```

### Промежуточный вывод.    
Вспомогательная метрика 1 - конверсия из первого депозита во второй - очень сильный рост, на 60.96%. p-value меньше alpha, изменения статистически значимы. Гипотеза роста подтверждена, конверсия выросла.    
Барьерная метрика  - конверсия из регистрации в первый депозит, упала на 2.1%, p-value - 0.939728 - не значимые изменения. Нулевая гипотеза подтверждена, конверсия не изменилась.

Начальная гипотеза подтверждена - пользователи, проходившие обучение, чаще возвращаются и их средний чек выше.

```python
# считаем ARPU и p-value по дням для каждой группы
df_sollmafin_abt['active_day'] = df_sollmafin_abt['event_ts'].dt.date
cumulative_arpu = df_sollmafin_abt.groupby(['group', 'active_day', 'user_id']).agg(
    total_spent = ('amount', 'sum')
).reset_index()

dates = cumulative_arpu['active_day'].sort_values().unique()

result_arpu = []
pval = []

for dt in dates:
    data = cumulative_arpu[cumulative_arpu['active_day'] <= dt]
    data = data.groupby(['group', 'user_id'])['total_spent'].sum().reset_index()
    
    group_mean = data.groupby('group')['total_spent'].mean().reset_index(name='arpu')
    result_arpu.append(group_mean.assign(active_day=dt))
    
    control = data[data['group'] == 'control']['total_spent']
    test = data[data['group'] == 'test']['total_spent']
    stat, p_value = ttest_ind(test, control, equal_var=False)
    pval.append({
        'active_day': dt,
        'p_value': p_value
    })    

cumulative_arpu_over_time = pd.concat(result_arpu)
pval_df = pd.DataFrame(pval)
```

Используя бутстрап, сравним разницы перцентилей (25, 50, 75) суммы всех депозитов на платящего пользователя в контрольной и тестовой группах:
- Построим доверительные интервалы для разницы между группами для каждого перцентиля.

```python
# уберем выбросы 99 перцентиля
control_paying_pct_99 = np.percentile(control_paying_pct, 99)
control_paying_filter = control_paying_pct[control_paying_pct <= control_paying_pct_99]

test_paying_pct_99 = np.percentile(test_paying_pct, 99)
test_paying_filter = test_paying_pct[test_paying_pct <= test_paying_pct_99]

print(
f'''
Количество платящих пользователей 
в контрольной группе после удаления выбросов: {control_paying_filter.shape[0]}

Количество платящих пользователей 
в тестовой группе после удаления выбросов: {test_paying_filter.shape[0]}
'''
)
```

```python
# бутстрап
# фиксируем сид
np.random.seed(1337) # LEET :)

n_iterations = 5000
prc = [25, 50, 75]

btstrp = {}

for percentil in prc:
    btstrp_diffs = []
    for i in range(n_iterations):
        btstrp_control = np.random.choice(
            control_paying_filter, 
            len(test_paying_filter),
            replace=True
        )
        btstrp_test = np.random.choice(
            test_paying_filter,
            len(test_paying_filter),
            replace=True
        )
        
        control_prc = np.percentile(btstrp_control, percentil)
        test_prc = np.percentile(btstrp_test, percentil)
        btstrp_diffs.append(test_prc - control_prc)
    
    btstrp_diffs = np.array(btstrp_diffs)
             
    # доверительный интервал
    lower = np.percentile(btstrp_diffs, alpha / 2 * 100)
    upper = np.percentile(btstrp_diffs, (1 - alpha / 2) * 100)
    
    btstrp[percentil] = {
        'diffs': btstrp_diffs,
        'ci_lower': lower,
        'ci_upper': upper,
        'abs_diff': np.percentile(test_paying_filter, percentil)
                    - np.percentile(control_paying_filter, percentil)
    }
    
    if btstrp[percentil]['ci_lower'] > 0:
        significance_message = (
            f"ДИ выше нуля, количество депозитов в {percentil} перцентиле выросло"
        )
    elif btstrp[percentil]['ci_upper'] < 0:
        significance_message = (
            f"ДИ ниже нуля, количество депозитов в {percentil} перцентиле снизилось"
        )
    else:
        significance_message = "Результат не значим"
    
    print(
        f'''
        {percentil} перцентиль:
        Контрольная группа: {np.percentile(control_paying_filter, percentil):.2f}
        Тестовая группа: {np.percentile(test_paying_filter, percentil):.2f}
        Доверительный интервал: [{lower:.2f}, {upper:.2f}]
        Абсолютная разница между группами: {btstrp[percentil]['abs_diff']:.2f}
        {significance_message}
        '''    
    )
```

```python
# визуализируем результат бутстрапа
plt.figure(figsize=(12, 6))
for percentil in prc:
    plt.hist(btstrp[percentil]['diffs'], bins=50, alpha=0.5, label=f'{percentil} перцентиль', edgecolor='black')
    plt.axvline(btstrp[percentil]['ci_lower'], color='red', linestyle='--', label=f'{percentil} перцентиль CI Lower')
    plt.axvline(btstrp[percentil]['ci_upper'], color='green', linestyle='--', label=f'{percentil} перцентиль CI Upper')
plt.title('Распределение разниц между группами для бутстрапа')
plt.xlabel('Разница')
plt.ylabel('Частота')
plt.legend()
plt.tight_layout()
plt.show()
```
